```python
from gensim.models import KeyedVectors

pretrained_model_path = r'D:\W2V\GoogleNews-vectors-negative300.bin.gz'  # Update with your model path
model = KeyedVectors.load_word2vec_format(pretrained_model_path, binary=True)
```

```python
word1 = 'cat'
word2 = 'dog'

similarity = model.similarity(word1, word2)
print(f"Similarity between '{word1}' and '{word2}' is: {similarity:.4f}")
```

```python
word_a = 'king'
word_b = 'man'
word_c = 'woman'

result = model.most_similar(positive=[word_b,word_c], negative=[word_a])
print(f"The result of analogy '{word_a} : {word_b} :: {word_c}:' is {result[0][0]}")
```

```python
import jieba
import re

def read_chinese_file_to_sentences(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
        text = re.sub(r'[^\s\d\w\u4e00-\u9fff]', '', text)
        text = re.sub(r'\s+', ' ', text)
        text = re.sub(r'\n', ' ', text)
    words = jieba.cut(text)
    sentences = [list(words)]
    return sentences
```

```python
import gensim
from gensim.models import Word2Vec
from gensim.models.word2vec import Word2Vec

file_path = './cnews.train.txt'
sentences = read_chinese_file_to_sentences(file_path)

model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4)

word_vectors = model.wv
model.save('cnews.model')
```

```python
loaded_model = Word2Vec.load('cnews.model')
similar_words = loaded_model.wv.most_similar('体育')

print("语义相似性：")
for word, similarity in similar_words:
    print(f"{word}: {similarity}")
```

```python
analogy_words = loaded_model.wv.most_similar(positive=['体育', '中国'], negative=['美国'], topn=10)

print(f"\n类比推理：体育 - 美国 + 中国：")
for word, analogy in analogy_words:
    print(f"{word}: {analogy:.4f}")
```
